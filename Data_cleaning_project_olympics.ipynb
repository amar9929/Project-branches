```python
year_filter= (Athletes_comp['year']<=2016) & (Athletes_comp['year']>1996)
```

```python
swimming_filter=Athletes_comp['sport']=='Swimming'
filter_400m=Athletes_comp['event'].str.startswith('400 metres Freestyle')
swim_loc=Athletes_comp[['year', 'sport', 'event', 'sex','medal', 'country']].loc[swimming_filter & filter_400m & year_filter]
swimming_pivot=pd.pivot_table(swim_loc, index=['year', 'country'],values='medal', aggfunc='count').sort_values(by=['year','medal'], ascending=False)

weightlift_filter=Athletes_comp['sport']=='Weightlifting'
filter_weight=Athletes_comp['event'].str.startswith('Heavyweight,')
weight_loc=Athletes_comp[['year', 'sport', 'event', 'sex','medal', 'country']].loc[weightlift_filter &  filter_weight & year_filter]
weight_pivot=pd.pivot_table(weight_loc, index=['year', 'country'], values='medal', aggfunc='count').sort_values(by=['year','medal'], ascending=False)

field_filter=Athletes_comp['sport']=='Athletics'
filter_pole=Athletes_comp['event'].str.startswith('Pole Vault,')
pole_loc=Athletes_comp[['year', 'sport', 'event', 'sex','medal', 'country']].loc[field_filter &  filter_pole & year_filter]
pole_pivot=pd.pivot_table(pole_loc, index=['year', 'country'], values='medal', aggfunc='count').sort_values(by=['year','medal'], ascending=False)



```
