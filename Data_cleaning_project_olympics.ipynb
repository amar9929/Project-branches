```python
bio=pd.read_csv('archive-4/Olympic_Athlete_Bio.csv')
```

```python
event=pd.read_csv('archive-4/Olympic_Athlete_Event_Results.csv')

```

```python
Athletes_comp=pd.merge(event,bio, how='left')
```

```python
Athletes_comp['year']=[int(i[0]) for i in event['edition'].str.split(' ')]
```

```python
Athletes_comp['medal'].replace({'na':None}, inplace=True)
```

```python
Athletes_comp['medal'].value_counts()
```

```python
Ath_medals=Athletes_comp[['year', 'sport', 'event', 'sex','medal', 'country']]
```

```python
year_filter= (Athletes_comp['year']<=2016) & (Athletes_comp['year']>=1996)
```

```python
swimming_filter=Athletes_comp['sport']=='Swimming'
filter_400m=Athletes_comp['event'].str.startswith('400 metres Freestyle')
swim_loc=Ath_medals.loc[swimming_filter & filter_400m & year_filter]
swim_test=pd.pivot_table(swim_loc, index='country', columns='year',values='medal', aggfunc='count', fill_value=0)
swim_pivot=swim_test.loc[(swim_test!=0).any(axis=1)].sort_values(by='country', ascending=False)
swim_pivot['Total']=swim_pivot.sum(axis=1)

weightlift_filter=Athletes_comp['sport']=='Weightlifting'
filter_weight=Athletes_comp['event'].str.startswith('Heavyweight,')
weight_loc=Ath_medals.loc[weightlift_filter &  filter_weight & year_filter]
weight_test=pd.pivot_table(weight_loc, index='country', columns='year',values='medal', aggfunc='count', fill_value=0)
weight_pivot=weight_test.loc[(weight_test!=0).any(axis=1)].sort_values(by='country', ascending=False)
weight_pivot['Total']=weight_pivot.sum(axis=1)

field_filter=Athletes_comp['sport']=='Athletics'
filter_pole=Athletes_comp['event'].str.startswith('Pole Vault,')
pole_loc=Ath_medals.loc[field_filter &  filter_pole & year_filter]
pole_test=pd.pivot_table(pole_loc, index='country', columns='year',values='medal', aggfunc='count', fill_value=0)#.sort_values(by=['year','medal'], ascending=False)
pole_pivot=pole_test.loc[(pole_test!=0).any(axis=1)].sort_values(by='country', ascending=False)
pole_pivot['Total']=pole_pivot.sum(axis=1)

```
